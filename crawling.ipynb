```python
from bs4 import BeautifulSoup as BS
import requests as req
```

```python
def get_posts_from_page(sp):
    posts_el = sp.find('div', {'id' : 'posts'})
    posts = {}
    for table in posts_el.find_all('table'):
        try:
            id_ = table['id']
            #print(id_)
            posts[id_] = table.prettify()
        except:
            pass
    print("Found {} posts".format(len(posts)))
    return posts
```

```python
# get all posts id's and html
url_start = "https://www.city-data.com/forum/health-wellness/3245374-have-you-had-covid-vaccine-side"
url_end = ".html"
posts_raw = {}
page_amount = 53
for i in range(1, 1+page_amount):
    print("Crawling page {} ...".format(i))
    url = "{}-{}{}".format(url_start, i, url_end)
    response = req.get(url)
    soup = BS(response.text, 'html.parser')
    page_posts = get_posts_from_page(soup)
    for k, v in page_posts.items():
        posts_raw[k] = v
print("Done! Got {} posts".format(len(posts_raw)))
```

```python
# save raw posts
import json
with open('posts_raw.json', 'w') as f:
    json.dump(posts_raw, f)
```
